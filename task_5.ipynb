```python
import numpy as np #for making arrays
import pandas as pd #it helps to make data frame(structure table)
from sklearn.model_selection import train_test_split #allows data to divide into training data and testing data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to panda dataframe
credit_card_data = pd.read_csv ('/content/credit caed.csv')
```

```python
# first five rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# data set info
credit_card_data.info()
```

```python
# checking the number of missinng values in each column
credit_card_data.isnull().sum()
```

```python
# check the disitribution of legit and fraudlent transaction
credit_card_data['Class'].value_counts()
```

The Data set is unbalanced

0--> normal trnsction 1--> fraudlent transaction

```python
# separating the data for analysis
legit= credit_card_data[credit_card_data.Class == 0]
fraud= credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of data
legit.Amount.describe()
```

```python
# statistical measures of data
fraud.Amount.describe()
```

Imp point to remember ***the mean of fraud transaction should be higer than the legit transactions

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

under-sampling --> used to dealing with unbalanced data

build a sample data set containing similar distribution of normal transaction and the fraudlent transaction

number of fraudlent transaction= 56

```python
legit_sample = legit.sample(n=56)
```

concatenating two Dataframes

```python
new_dataset =pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

split the data into features and targets

```python
x= new_dataset.drop(columns='Class' , axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

split data into training data and testing data

```python
x_train, x_test , y_train, y_test = train_test_split(x,y, test_size=0.2 , stratify= y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

model training

logistic regression

```python

model= LogisticRegression(max_iter=100000)
```

```python
# training the logistic Regression model with training data
model.fit(x_train,y_train)
```

model evaluation

```python
# accuracy score on training data
x_train_prediction= model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction, y_train)
```

```python
print('accuracy on data :', training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction= model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction, y_test)
```

```python
print('accuracy score on test data :' , test_data_accuracy)
```
